```python
import nltk
from collections import defaultdict
from nltk.util import ngrams

# 1. تحميل الموارد المطلوبة
nltk.download('gutenberg', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True) # السطر ده هو اللي هيحل الـ Error

# 2. تجهيز الداتا
raw_text = nltk.corpus.gutenberg.raw('austen-emma.txt')[:50000]
tokens = nltk.word_tokenize(raw_text.lower())

# 3. بناء الموديل (Bi-gram Model مع Laplace Smoothing)
model = defaultdict(lambda: defaultdict(lambda: 0.001))
bigrams = list(ngrams(tokens, 2))

for w1, w2 in bigrams:
    model[w1][w2] += 1

# تحويل التكرارات لاحتمالات (Normalization)
for w1 in model:
    total_count = sum(model[w1].values())
    for w2 in model[w1]:
        model[w1][w2] /= total_count

# 4. دالة التوقع
def get_suggestions(word, top_n=5):
    word = word.lower()
    if word in model:
        suggestions = sorted(model[word].items(), key=lambda item: item[1], reverse=True)
        return [item[0] for item in suggestions[:top_n]]
    return ["لا توجد اقتراحات حالياً"]

# تجربة الموديل
test_word = "emma"
print(f"✅ الموديل جاهز! اقتراحات لكلمة '{test_word}':")
print(get_suggestions(test_word))
```

```python
import streamlit as st
import nltk
from collections import defaultdict
from nltk.util import ngrams

# إعدادات الصفحة
st.set_page_config(page_title="AI Autofill", page_icon="📝")

@st.cache_resource
def load_model():
    nltk.download('gutenberg', quiet=True)
    nltk.download('punkt_tab', quiet=True)
    raw_text = nltk.corpus.gutenberg.raw('austen-emma.txt')
    tokens = [w.lower() for w in nltk.word_tokenize(raw_text) if w.isalpha()]

    # استخدام Trigrams (سياق كلمتين لتوقع الكلمة الثالثة)
    model = defaultdict(lambda: defaultdict(lambda: 0))
    for w1, w2, w3 in ngrams(tokens, 3):
        model[(w1, w2)][w3] += 1

    # تحويل التكرارات لاحتمالات
    for context in model:
        total = sum(model[context].values())
        for w3 in model[context]:
            model[context][w3] /= total
    return model

model = load_model()

st.title("🚀 Smart Autofill System")
input_text = st.text_input("اكتب جملة (على الأقل كلمتين):", "")

# ميزة الـ Auto-update: تظهر الاقتراحات فور الكتابة
if input_text:
    words = input_text.lower().split()
    if len(words) >= 2:
        context = (words[-2], words[-1])
        if context in model:
            suggestions = sorted(model[context].items(), key=lambda x: x[1], reverse=True)[:3]
            st.write("### الكلمات المتوقعة:")
            for word, prob in suggestions:
                st.button(f"{word} ({prob:.2%})")
        else:
            st.warning("لا توجد اقتراحات لهذا السياق.")
    else:
        st.info("اكتب كلمتين على الأقل لبدء التوقع...")
```

```python
import streamlit as st
import nltk
from collections import defaultdict
from nltk.util import ngrams

# 1. إعدادات الصفحة الاحترافية
st.set_page_config(page_title="AI Autofill System", page_icon="📝")

# 2. تنسيق CSS لمظهر عصري (بصفتك Front-end Dev)
st.markdown("""
    <style>
    .stButton>button {
        width: 100%;
        border-radius: 20px;
        background-color: #f0f2f6;
        border: 1px solid #d1d1d1;
    }
    .stButton>button:hover {
        background-color: #4CAF50;
        color: white;
    }
    </style>
""", unsafe_allow_html=True)

# 3. تحميل الموديل (تخزين مؤقت)
@st.cache_resource
def load_model():
    nltk.download('gutenberg', quiet=True)
    nltk.download('punkt_tab', quiet=True)
    raw_text = nltk.corpus.gutenberg.raw('austen-emma.txt')
    tokens = [w.lower() for w in nltk.word_tokenize(raw_text) if w.isalpha()]

    # بناء Trigram Model
    model = defaultdict(lambda: defaultdict(lambda: 0))
    for w1, w2, w3 in ngrams(tokens, 3):
        model[(w1, w2)][w3] += 1

    # حساب الاحتمالات
    for context in model:
        total = sum(model[context].values())
        for w3 in model[context]:
            model[context][w3] /= total
    return model

model = load_model()

# 4. واجهة المستخدم
st.title("🚀 Smart Autofill System")
st.write("نظام التوقع الذكي يعتمد على سياق الكلمتين السابقتين.")

# استخدام Session State للحفاظ على النص المكتوب
if 'text' not in st.session_state:
    st.session_state.text = ""

input_text = st.text_input("اكتب الجملة هنا:", value=st.session_state.text)

# 5. منطق التوقع
if input_text:
    words = input_text.lower().split()
    if len(words) >= 2:
        context = (words[-2], words[-1])
        if context in model:
            # ترتيب الاقتراحات
            suggestions = sorted(model[context].items(), key=lambda x: x[1], reverse=True)[:3]

            st.write("---")
            st.subheader("💡 كلمات مقترحة:")
            cols = st.columns(3)

            for i, (word, prob) in enumerate(suggestions):
                with cols[i]:
                    if st.button(f"{word.capitalize()}"):
                        # عند الضغط على الزر، يتم إضافة الكلمة للنص
                        st.session_state.text = input_text + " " + word
                        st.rerun() # تحديث الصفحة فوراً لإظهار الكلمة الجديدة
        else:
            st.warning("عذراً، لا توجد اقتراحات لهذا السياق.")
    else:
        st.info("استمر في الكتابة (نحتاج كلمتين على الأقل)...")
```
